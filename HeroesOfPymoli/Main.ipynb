```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
csv_path = "../Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
heroes_df = pd.read_csv(csv_path)
heroes_df
```

```python
#Listing object types
heroes_df.dtypes
```

```python
# Player Count
PlayerNumber = heroes_df.drop_duplicates(subset=['SN','Gender','Age'])
PlayerTotalNumber = len(PlayerNumber)
df = pd.DataFrame({"Total_Players":[PlayerTotalNumber]})
df
```

```python
#UniqueItems = heroes_df.drop_duplicates(subset=['Item ID'])
#ItemTotalNumber = len(UniqueItems)

#Purchasing Analysis (Total)
itemcount= len(heroes_df["Item ID"].unique())
average_price= heroes_df["Price"].mean()
total_purchases= len(heroes_df["Purchase ID"])
total_revenue= sum(heroes_df["Price"])

df = pd.DataFrame({"Number_of_Unique_Items":[itemcount],"Average_Price":[average_price],
                   "Number_of_Purchases":[total_purchases],"Total_Revenue":[total_revenue]})
df
#itemcount
```

```python
#Gender Demographics
gender_count = PlayerNumber["Gender"].value_counts()

gender_percentage = (gender_count/PlayerTotalNumber)*100

df2 = pd.DataFrame({"Total_Count":gender_count,"Percentage_of_Players":gender_percentage})
df2
```

```python
#Purchases sorted by each gender
gender_purchasing_data = heroes_df.groupby("Gender")
gender_purchasing_total = gender_purchasing_data["Gender"].count()

#Averages
gender_purchasing_quantity_average = ((gender_purchasing_data["Price"].count()/len("Purchase ID")))
gender_purchasing_price_average = gender_purchasing_data["Price"].mean()

#Total revenue
gender_purchasing_value = gender_purchasing_data["Price"].sum()


df4= pd.DataFrame({"Number_Of_Purchases_By_Gender":gender_purchasing_total,"Average_Purchase_Total_Per_Person_By_Gender":
                   gender_purchasing_quantity_average,"Average_Purchase_Price_By_Gender":
                   gender_purchasing_price_average,"Total_Revenue_By_Gender":gender_purchasing_value})
df4
```

```python
#Age Demographics bin set up

bins = [0, 9, 14, 19, 24, 29, 34, 39, 111]

group_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

heroes_df["Age Brackets"] = pd.cut(heroes_df["Age"], bins, labels=group_labels)
age_bracket_df = heroes_df.groupby(["Age Brackets"])


ages_list = age_bracket_df["SN"].nunique()

age_brackets_df = pd.DataFrame({"Total_By_Bracket": ages_list})
age_brackets_df
```

```python
#Age Demographics
purchase_list = age_bracket_df["Purchase ID"].count()
average_purchasing_price = age_bracket_df["Price"].mean()
total_purchasing_value = age_bracket_df["Price"].sum()
#average_price_per_person

df5 = pd.DataFrame({"Purchase_Count": purchase_list, "Average_Purchase_Price": average_purchasing_price,
                    "Total_Purchase_Value": total_purchasing_value})
df5
```

```python
#Top Spenders
individual_spenders = heroes_df.groupby(["SN"])
individual_spenders

purchases_total = individual_spenders["Purchase ID"].count()
average_spender_purchase_price = individual_spenders["Price"].mean()
total_purchase_spender_value = individual_spenders["Price"].sum()


df6 = pd.DataFrame({"Purchase_Count": purchases_total, "Average_Purchase_Price": average_spender_purchase_price,
                    "Total_Purchase_Value": total_purchase_spender_value})
df6 = df6.sort_values(["Total_Purchase_Value"], ascending=False)

df6[:5]
```

```python
#Most Popular Items
items_list = heroes_df[["Item ID","Item Name","Price"]]

grouped_items_list = items_list.groupby(["Item ID", "Item Name"])

purchase_count = grouped_items_list["Item ID"].count()
item_price = grouped_items_list["Price"].mean()
total_purchase_value = grouped_items_list["Price"].sum()

df7 = pd.DataFrame({"Purchase_Count": purchase_count,"Item_Price": item_price,
                    "Total_Purchase_Value": total_purchase_value})
df7 = df7.sort_values(["Purchase_Count"], ascending=False)
df7[:5]
```

```python
#Most Profitable Items
items_list2 = heroes_df[["Item ID","Item Name","Price"]]

grouped_items_list2 = items_list2.groupby(["Item ID", "Item Name"])

purchase_count = grouped_items_list2["Item ID"].count()
item_price = grouped_items_list2["Price"].mean()
total_purchase_value = grouped_items_list2["Price"].sum()

df8 = pd.DataFrame({"Purchase_Count": purchase_count,"Item_Price": item_price,
                    "Total_Purchase_Value": total_purchase_value})
df8 = df8.sort_values(["Total_Purchase_Value"], ascending=False)
df8[:5]
```
